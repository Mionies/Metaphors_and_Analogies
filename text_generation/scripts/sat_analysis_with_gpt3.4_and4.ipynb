```python
#['predicted', 'idx', 'true', 'class_metrics']
def satstar(idx, predicted,trues):
    #print(len(idx),len(predicted),len(trues))
    dres={1:[0,0],2:[0,0]}
    for i,p in enumerate(idx):
        if p in didx2label:
            if didx2label[p]==1:
                dres[1][1]+=1
            elif didx2label[p]==2:
                dres[2][1]+=1
            if predicted[i]==trues[i]:
                if didx2label[p]==1:
                    dres[1][0]+=1
                elif didx2label[p]==2:
                    dres[2][0]+=1
    return dres
                
    #a[expe][]
```

```python
res = {}
for one_run in sat_permutation:
    if not one_run["model"]+str(one_run["prompt"]) in res:
        res[one_run["model"]+str(one_run["prompt"]) ]={1:[0,0],2:[0,0]}
    dres= satstar(one_run["idx"],one_run["predicted"],one_run["true"])
    res[one_run["model"]+str(one_run["prompt"])][1][0]+=dres[1][0]
    res[one_run["model"]+str(one_run["prompt"])][1][1]+=dres[1][1]
    res[one_run["model"]+str(one_run["prompt"])][2][0]+=dres[2][0]
    res[one_run["model"]+str(one_run["prompt"])][2][1]+=dres[2][1]

k = []
for x in res:    
    k.append([x,'\t', round(res[x][1][0]*100/res[x][1][1],1),"\t",round(res[x][2][0]*100/res[x][2][1],1)])

k.sort(key = lambda w:w[0])

for x in k:
    print("\t".join([str(y) for y in x]))
```

```python
def satstarsign(idx, predicted,trues):
    dres={1:[],2:[]}
    for i,p in enumerate(idx):
        if p in didx2label:
            if predicted[i]==trues[i]:
                if didx2label[p]==1:
                    dres[1].append(1)
                elif didx2label[p]==2:
                    dres[2].append(1)
            else:
                if didx2label[p]==1:
                    dres[1].append(0)
                elif didx2label[p]==2:
                    dres[2].append(0)
    return dres
```

```python
import statistics
import scipy
import numpy as np

ressign = {}
for one_run in sat_permutation:
    if not one_run["model"]+str(one_run["prompt"]) in ressign:
        ressign[one_run["model"]+str(one_run["prompt"]) ]={1:[],2:[]}
    dres= satstarsign(one_run["idx"],one_run["predicted"],one_run["true"])
    ressign[one_run["model"]+str(one_run["prompt"])][1].extend(dres[1])
    ressign[one_run["model"]+str(one_run["prompt"])][2].extend(dres[2])
  

k = []
for x in ressign:    
    accuracy_sig = round(scipy.stats.ttest_ind(ressign[x][1],ressign[x][2],equal_var = False).pvalue,4)
    p005 = accuracy_sig<0.05
    k.append([x,'\t', round(sum(ressign[x][1])*100/len(ressign[x][1]),1),"\t",round(sum(ressign[x][2])*100/len(ressign[x][2]),1),accuracy_sig,p005])

k.sort(key = lambda w:w[0])

for x in k:
    print("\t".join([str(y) for y in x]))
```

```python
import os
import json
from datasets import load_dataset
from sklearn.metrics import (
                          f1_score,
                          precision_recall_fscore_support,
                          accuracy_score,
                          balanced_accuracy_score,
                          precision_score,
                          recall_score,
)

def tosetid(i):
    return int(i/5)

def EvaluationMetrics(y_pred,y_true):
    analysis={}
    #analysis["logits"] = [[y*1 for y in x] for x in logits]
    analysis["accuracy"]={}
    #micro
    analysis["micro-f1"]= round(f1_score(y_true,y_pred,average="micro"),4)
    analysis["micro-precision"]=round(precision_score(y_true,y_pred,average="micro"),4)
    analysis["micro-recall"]=round(recall_score(y_true,y_pred, average="micro"),4)
    # macro
    analysis["macro-f1"]=round(f1_score(y_true,y_pred,average="macro"),4)
    analysis["macro-precision"]=round(precision_score(y_true,y_pred,average="macro"),4)
    analysis["macro-recall"]=round(recall_score(y_true,y_pred,average="macro"),4)
    #weighted
    analysis["weighted-f1"]=round(f1_score(y_true,y_pred,average="weighted"),4)
    analysis["weighted-precision"]=round(precision_score(y_true,y_pred,average="weighted"),4)
    analysis["weighted-recall"]=round(recall_score(y_true,y_pred,average="weighted"),4)
    #Accuracy
    analysis["accuracy"]["balanced"]=round(balanced_accuracy_score(y_true,y_pred),4)
    analysis["accuracy"]["standard"]=round(accuracy_score(y_true,y_pred),4)
    analysis["class_prf1"]=[list(x) for x in precision_recall_fscore_support(y_true,y_pred)]
    for i,x in enumerate(analysis["class_prf1"][:-1]):
        analysis["class_prf1"][i] = [round(y,4) for y in x]
    analysis["class_prf1"][-1]= [int(x*1) for x in analysis["class_prf1"][-1 ]]
    return analysis
```

```python
from statistics import mean

def Group(list_v,n):
    m = len(list_v)/n
    g = []
    for x in range(int(m)):
        g.append([])
    for i,x in enumerate(list_v):
        k = int(i/5)
        g[k].append(x)
    return g

means = []
labels = Group(a["davinci"]["true"],5)  
total_n_label=[a[model]["true"].count(i) for i in range(3)]


final = []
for model in a:
    per_class = [[],[],[]]
    for i,x in enumerate(a[model]["score"]):
        per_class[a[model]["true"][i]].append(x)
    means.append([model, [mean(y) for y in per_class]])
    results = []
    
    scores = Group(a[model]["score"],5)  
    for i,x in enumerate(scores):
        one_set = list(zip(x,labels[i]))
        one_set.sort(key=lambda w:w[0])
        results.append(one_set[0][1])
    n_correct = [round(results.count(i)*100/total_n_label[i],1) for i in [1,2]]
    final.append([model,n_correct[0],n_correct[1]])

final.sort()

for x in final:
    print("\t".join([str(y) for y in x]))
        
```
